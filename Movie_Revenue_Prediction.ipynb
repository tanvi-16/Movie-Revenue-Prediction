```python
from datetime import timedelta
def convert_running_time(time_str):
    if pd.isna(time_str):  
        return None
    
    time_parts = time_str.split()
    hours = 0
    minutes = 0

    if "hr" in time_parts:
        hours = int(time_parts[time_parts.index("hr") - 1])
    if "min" in time_parts:
        minutes = int(time_parts[time_parts.index("min") - 1])

    return str(timedelta(hours=hours, minutes=minutes))

df["Running Time"] = df["Running Time"].apply(convert_running_time)
```

```python
from datetime import timedelta
# Convert 'Running Time' from HH:MM:SS or MM:SS to total minutes
def convert_runtime(runtime):
    try:
        if isinstance(runtime, str):
            parts = runtime.split(':')
            if len(parts) == 3:  # Format HH:MM:SS
                hours, minutes, _ = map(int, parts)
                return hours * 60 + minutes
            elif len(parts) == 2:  # Format MM:SS
                minutes, _ = map(int, parts)
                return minutes
        return float(runtime)  # If it's already numeric
    except:
        return np.nan  # If conversion fails, return NaN

df['Running Time'] = df['Running Time'].astype(str).apply(convert_runtime)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from skopt import BayesSearchCV  # Bayesian Optimization
```

```python
# Feature Engineering
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df['Release Month'] = df['Release Date'].dt.month
df['Release Year'] = df['Release Date'].dt.year
df['Holiday Season'] = df['Release Month'].apply(lambda x: 1 if x in [11, 12, 6, 7] else 0)
df['Log Revenue'] = np.log1p(df['Worldwide'])  # Log transform for skewed revenue
```

```python
# Adding new features
df['Has Sequel'] = df['Name'].apply(lambda x: 1 if any(seq in x for seq in ['2', '3', 'II', 'III', 'IV']) else 0)
df['Franchise'] = df['Name'].apply(lambda x: 1 if any(franchise in x for franchise in ['Avengers', 'Harry Potter', 'Star Wars', 'Batman', 'Spider-Man']) else 0)
```

```python
# Avoiding Data Leakage for Popularity Features
director_revenue = df.groupby('Director')['Worldwide'].mean()
lead_actor_revenue = df.groupby('Cast')['Worldwide'].mean()
df['Director Popularity'] = df['Director'].map(director_revenue)
df['Lead Actor Popularity'] = df['Cast'].map(lead_actor_revenue)
```

```python
# Multi-Hot Encoding for Genres
df = df.join(df['Genres'].str.get_dummies(sep=', '))
```

```python
# Selecting Features & Target Variables
features = ['Budget', 'Running Time', 'Release Month', 'Release Year', 'Holiday Season', 'Has Sequel', 'Franchise', 'Director Popularity', 'Lead Actor Popularity'] + list(df['Genres'].str.get_dummies(sep=', ').columns)
target = 'Log Revenue'

df = df[features + [target]].dropna()
```

```python
# Preprocessing Pipelines
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, features)
])
```

```python
# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
# Model Training & Evaluation
def train_evaluate(model, X_train, X_test, y_train, y_test):
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='r2')
    cv_r2 = np.mean(cv_scores)
    
    print(f"{model.__class__.__name__} RMSE: {rmse:.2f}, R²: {r2:.2f}, CV R²: {cv_r2:.2f}")
    
    return pipeline
```

```python
# Base Models
rf_model = train_evaluate(RandomForestRegressor(n_estimators=500, max_depth=30, random_state=42), X_train, X_test, y_train, y_test)
xgb_model = train_evaluate(XGBRegressor(n_estimators=500, learning_rate=0.03, max_depth=8, enable_categorical=True, random_state=42), X_train, X_test, y_train, y_test)
```

```python
# Stacking Model
stacked_model = StackingRegressor(
    estimators=[
        ('rf', RandomForestRegressor(n_estimators=500, max_depth=30, random_state=42)),
        ('xgb', XGBRegressor(n_estimators=500, learning_rate=0.03, max_depth=8, enable_categorical=True, random_state=42)),
        ('lgbm', LGBMRegressor(n_estimators=500, learning_rate=0.03, max_depth=8, random_state=42)),
        ('catboost', CatBoostRegressor(n_estimators=500, learning_rate=0.03, max_depth=8, random_state=42, verbose=0))
    ],
    final_estimator=GradientBoostingRegressor(n_estimators=300, learning_rate=0.03, max_depth=5, random_state=42)
)

stacked_model = train_evaluate(stacked_model, X_train, X_test, y_train, y_test)
```

```python
# Bayesian Optimization for XGBoost
param_space = {
    'n_estimators': (500, 1000),
    'learning_rate': (0.01, 0.1, 'log-uniform'),
    'max_depth': (6, 12),
    'subsample': (0.7, 0.9),
    'colsample_bytree': (0.7, 0.9),
    'reg_alpha': (0, 1),
    'reg_lambda': (1, 10)
}

opt_xgb = BayesSearchCV(
    XGBRegressor(enable_categorical=True, random_state=42),
    param_space,
    n_iter=50,  # Increased optimization steps
    cv=5,
    scoring='r2',
    n_jobs=-1
)

opt_xgb.fit(X_train, y_train)
print(f"Best XGB Params: {opt_xgb.best_params_}, Best R²: {opt_xgb.best_score_:.2f}")
```

```python
import shap
# SHAP Importance
explainer = shap.Explainer(opt_xgb.best_estimator_)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)
```

```python
def predict_movie_revenue(model, df):
    try:
        movie_df = pd.DataFrame([df])
        
        # Apply the same preprocessing as training data
        movie_transformed = preprocessor.transform(movie_df)
        
        predicted_log_revenue = model.predict(movie_transformed)[0]
        return np.expm1(predicted_log_revenue)  # Convert log revenue back to normal scale
    except Exception as e:
        print("Error in prediction:", e)
        return None
```

```python
sample_movie = {
    'Budget': 20000000,  # $200M
    'Running Time': 150,  # 150 minutes
    'Release Month': 6,  # June (summer blockbuster)
    'Release Year': 2025,
    'Holiday Season': 1,  # Summer holiday
    'Has Sequel': 1,  # Part of a franchise
    'Franchise': 1,  # Recognized franchise
    'Director Popularity': 800000000,  # Previous avg revenue $800M
    'Lead Actor Popularity': 600000000,  # Previous avg revenue $600M
}

# Add genre flags (e.g., 'Action', 'Adventure', 'Sci-Fi' etc.)
genres = ['Action', 'Adventure', 'Sci-Fi', 'Drama']
for genre in df.columns:
    if genre in genres:
        sample_movie[genre] = 1
    elif genre not in sample_movie:
        sample_movie[genre] = 0  # Ensure all required features exist

# Predict Revenue
predicted_revenue = predict_movie_revenue(opt_xgb.best_estimator_, sample_movie)
print(f"Predicted Revenue: ${predicted_revenue:,.2f}")
```
